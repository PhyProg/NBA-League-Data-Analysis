```python
import numpy as np
import pandas as pd
import tensorflow as tf

from matplotlib import pyplot as plt

from draft_data import *
from draft_func import *
```

## **Plot all obtainable data, saving plots**

### Plot yearly WS 

```python
for year in range(1978, 2015):
    ws, yr = draft_yearly_WS(year)
    
    fig, ax = plt.subplots()
    ax.scatter(yr, ws, c = 'r', edgecolors = 'k', label = str(i))
    ax.plot(yr, ws, c = 'b')
    ax.set_title('Draft'+str(year)+'yearly WS')
    ax.set_ylabel('WS')
    ax.set_xlabel('Season')
    ax.grid(True, color = 'gray', linestyle = 'dashed')
    
    plt.show()
    fig.savefig('../../data/draft/plots/d_yearly_ws_'+str(year)+'.PNG')
    plt.close(fig)
```

### Total WS per draft year

```python
draft = np.arange(1978, 2016, 1)
ws = np.arange(1978, 2016, 1)
for year in draft:
    ws[year-1978] = draft_total_WS(year)
    
fig, ax = plt.subplots()
ax.scatter(draft, ws, c = 'r', edgecolors = 'k')
ax.plot(draft, ws, c = 'b')
ax.set_title('WS vs Draft year')
ax.set_ylabel('WS')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/WS_per_draft.PNG')
plt.close(fig)
```

### Total years of players in a league versus draft year

```python
draft = np.arange(1978, 2016, 1)
yr = np.arange(1978, 2016, 1.00)
for year in draft:
    yr[year-1978] = draft_total_years(year)
    
fig, ax = plt.subplots()
ax.scatter(draft, yr, c = 'r', edgecolors = 'k')
ax.plot(draft, yr, c = 'b')
ax.set_title('Total years vs Draft year')
ax.set_ylabel('Total years')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/years_per_draft.PNG')
plt.close(fig)
```

### Average years spent by players in a league versus draft year

```python
draft = np.arange(1978, 2016, 1)
yr = np.arange(1978, 2016, 1.00)
for year in draft:
    yr[year-1978] = draft_avg_years(year)
    
fig, ax = plt.subplots()
ax.scatter(draft, yr, c = 'r', edgecolors = 'k')
ax.plot(draft, yr, c = 'b')
ax.set_title('Average years vs Draft year')
ax.set_ylabel('Average years')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/avg_years_per_draft.PNG')
plt.close(fig)
```

### Average WS per player per draft class

```python
pl_avg_years = np.arange(1978, 2016, 1)
pl_tot_years = np.arange(1978, 2016, 1)
pl_num = np.arange(1978, 2016, 1)
for i in range(len(draft_class)):
    pl_num[i] = len(draft_class[i])
    pl_avg_years[i] = draft_avg_years(i+1978)
    pl_tot_years[i] = draft_total_years(i+1978)
    
#print(pl_num, pl_avg_years, pl_tot_years)
draft = np.arange(1978, 2016, 1)
ws = np.arange(1978, 2016, 1)
for year in draft:
    ws[year-1978] = draft_total_WS(year)
    
fig, ax = plt.subplots()
ax.scatter(draft, ws/pl_num, c = 'r', edgecolors = 'k')
ax.plot(draft, ws/pl_num, c = 'b')
ax.set_title('WS/nop vs Draft year')
ax.set_ylabel('WS/nop')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/WS_pnop_per_draft.PNG')
plt.close(fig)

fig, ax = plt.subplots()
ax.scatter(draft, ws/pl_avg_years, c = 'r', edgecolors = 'k')
ax.plot(draft, ws/pl_avg_years, c = 'b')
ax.set_title('WS/ay vs Draft year')
ax.set_ylabel('WS/ay')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/WS_pay_per_draft.PNG')
plt.close(fig)

fig, ax = plt.subplots()
ax.scatter(draft, ws/pl_tot_years, c = 'r', edgecolors = 'k')
ax.plot(draft, ws/pl_tot_years, c = 'b')
ax.set_title('WS/ty vs Draft year')
ax.set_ylabel('WS/ty')
ax.set_xlabel('Draft year')
ax.grid(True, color = 'gray', linestyle = 'dashed')
    
plt.show()
fig.savefig('../../data/draft/plots/WS_pty_per_draft.PNG')
plt.close(fig)
```
